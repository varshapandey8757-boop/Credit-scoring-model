```python
# ─────────────────────────────────────────────────────────
# STEP 0: INSTALL / IMPORT LIBRARIES
# ─────────────────────────────────────────────────────────
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

# Machine Learning
from sklearn.model_selection import train_test_split, cross_val_score, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

```python
#algorithms
from sklearn.linear_model import LogisticRegression   # add this line
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

lr = LogisticRegression(...)   # now this works
```

```python
# Metrics
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score, roc_curve,
    confusion_matrix, classification_report
)

print("=" * 65)
print("   CREDIT SCORE MODEL — Bank Loan Default Prediction")
print("=" * 65)
```

```python
# ─────────────────────────────────────────────────────────
# STEP 1: LOAD DATASET
# ─────────────────────────────────────────────────────────
print("\n[STEP 1] Loading Dataset...")

# ⚠️ Update this path if running locally
df = pd.read_csv("Bankloan_Cleanedv1.csv")

print(f"  ✓ Dataset loaded: {df.shape[0]} rows × {df.shape[1]} columns")
print(f"\n  Columns: {list(df.columns)}")
print(f"\n  First 5 rows:\n{df.head()}")
print(f"\n  Data Types:\n{df.dtypes}")
print(f"\n  Missing Values:\n{df.isnull().sum()}")
print(f"\n  Class Distribution:\n{df['default'].value_counts()}")
print(f"\n  Default Rate : {df['default'].mean()*100:.1f}%")
print(f"  Non-Default  : {(1-df['default'].mean())*100:.1f}%")

```

```python
# ─────────────────────────────────────────────────────────
# STEP 2: DATA PREPROCESSING
# ─────────────────────────────────────────────────────────
print("\n[STEP 2] Preprocessing Data...")

# 2.1 Separate features and target
X = df.drop("default", axis=1)
y = df["default"]

print(f"  Features (X): {list(X.columns)}")
print(f"  Target   (y): 'default'  →  0 = No Default, 1 = Default")

# 2.2 Handle missing values — median imputation
#     Fit imputer on training data only (no data leakage)
imputer = SimpleImputer(strategy="median")
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)
print(f"\n  ✓ Missing values imputed with column medians")
print(f"  Missing after imputation: {X_imputed.isnull().sum().sum()}")
```

```python
# 2.3 Train/Test Split (80/20 stratified)
X_train, X_test, y_train, y_test = train_test_split(
    X_imputed, y,
    test_size=0.2,
    random_state=42,
    stratify=y          # preserve class ratio in both sets
)
print(f"\n  ✓ Train/Test Split (80/20 stratified):")
print(f"     Training set : {X_train.shape[0]} records  "
      f"(Default: {y_train.sum()}, No Default: {(y_train==0).sum()})")
print(f"     Test set     : {X_test.shape[0]} records  "
      f"(Default: {y_test.sum()}, No Default: {(y_test==0).sum()})")
```

```python
# 2.4 Feature Scaling — required for Logistic Regression & SVM
scaler = StandardScaler()
X_train_sc = scaler.fit_transform(X_train)   # fit on train only
X_test_sc  = scaler.transform(X_test)         # transform test with same scaler
X_all_sc   = scaler.fit_transform(X_imputed)  # for cross-validation
print(f"\n  ✓ StandardScaler applied (mean=0, std=1)")
print(f"     Used for: Logistic Regression and SVM")
print(f"     NOT used for: Decision Tree and Random Forest (not needed)")
```

```python
# ─────────────────────────────────────────────────────────
# STEP 3: FEATURE ENGINEERING
# ─────────────────────────────────────────────────────────
print("\n[STEP 3] Feature Engineering Insights...")

print("""
  Feature Summary (Business Meaning):
  ┌────────────┬────────────────────────────────────────────────────────┐
  │ Feature    │ Business Meaning                                       │
  ├────────────┼────────────────────────────────────────────────────────┤
  │ age        │ Applicant age (years) — older = more credit history   │
  │ employ     │ Years at current employer — stability indicator        │
  │ address    │ Years at current address — residential stability       │
  │ income     │ Annual income ($000s) — repayment capacity            │
  │ debtinc    │ Debt-to-income ratio (%) — KEY risk metric            │
  │ creddebt   │ Credit card debt ($000s) — revolving debt burden      │
  │ othdebt    │ Other debts ($000s) — total liability                 │
  │ ed         │ Education level (1–5 scale)                           │
  │ default    │ TARGET: 1=defaulted, 0=did not default                │
  └────────────┴────────────────────────────────────────────────────────┘

  Engineered Insights:
  • Total Debt       = creddebt + othdebt
  • Debt Coverage    = income / (creddebt + othdebt)
  • debtinc > 40%    → High risk zone (industry benchmark)
  • employ < 2 years → Unstable income risk
""")
```

```python
# ─────────────────────────────────────────────────────────
# STEP 4: DEFINE & TRAIN ALL 4 MODELS
# ─────────────────────────────────────────────────────────
print("\n[STEP 4] Training Classification Models...")

# Cross-validation setup
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

```python
# ── 4.1 Logistic Regression ──────────────────────────────
print("\n  ── Model 1: Logistic Regression ──")
lr = LogisticRegression(
    C=1.0,                     # Inverse regularization strength
    class_weight="balanced",   # Handle class imbalance
    max_iter=1000,             # Allow enough iterations
    random_state=42
)
lr.fit(X_train_sc, y_train)
lr_pred  = lr.predict(X_test_sc)
lr_proba = lr.predict_proba(X_test_sc)[:, 1]
lr_cv    = cross_val_score(lr, X_all_sc, y, cv=cv, scoring="roc_auc")
print(f"  ✓ Logistic Regression trained")
print(f"    Coefficients: {dict(zip(X.columns, lr.coef_[0].round(3)))}")
```

```python
# ── 4.2 Decision Tree ────────────────────────────────────
print("\n  ── Model 2: Decision Tree ──")
dt = DecisionTreeClassifier(
    max_depth=6,               # Limit depth to prevent overfitting
    criterion="gini",          # Gini impurity for splits
    class_weight="balanced",
    random_state=42
)
dt.fit(X_train, y_train)
dt_pred  = dt.predict(X_test)
dt_proba = dt.predict_proba(X_test)[:, 1]
dt_cv    = cross_val_score(dt, X_imputed, y, cv=cv, scoring="roc_auc")
print(f"  ✓ Decision Tree trained  (depth={dt.get_depth()}, leaves={dt.get_n_leaves()})")
print(f"\n  Top Decision Rules:")
print(export_text(dt, feature_names=list(X.columns), max_depth=3))
```

```python
# ── 4.3 Random Forest ────────────────────────────────────
print("\n  ── Model 3: Random Forest ──")
rf = RandomForestClassifier(
    n_estimators=200,          # Number of trees
    max_depth=10,              # Max depth per tree
    min_samples_split=5,       # Min samples to split node
    class_weight="balanced",
    n_jobs=-1,                 # Use all CPU cores
    random_state=42
)
rf.fit(X_train, y_train)
rf_pred  = rf.predict(X_test)
rf_proba = rf.predict_proba(X_test)[:, 1]
rf_cv    = cross_val_score(rf, X_imputed, y, cv=cv, scoring="roc_auc")
print(f"  ✓ Random Forest trained  ({rf.n_estimators} trees)")
```

```python
# ── 4.4 SVM ──────────────────────────────────────────────
print("\n  ── Model 4: SVM (RBF Kernel) ──")
svm = SVC(
    kernel="rbf",              # Radial Basis Function kernel
    C=10,                      # High penalty = fewer misclassifications
    gamma="scale",             # gamma = 1/(n_features * X.var())
    class_weight="balanced",
    probability=True,          # Enable predict_proba
    random_state=42
)
svm.fit(X_train_sc, y_train)
svm_pred  = svm.predict(X_test_sc)
svm_proba = svm.predict_proba(X_test_sc)[:, 1]
svm_cv    = cross_val_score(svm, X_all_sc, y, cv=cv, scoring="roc_auc")
print(f"  ✓ SVM trained  (kernel=RBF, C={svm.C}, support_vectors={svm.n_support_.sum()})")

```

```python
# ─────────────────────────────────────────────────────────
# STEP 5: EVALUATION METRICS
# ─────────────────────────────────────────────────────────
print("\n[STEP 5] Evaluating All Models...\n")

def evaluate_model(name, y_true, y_pred, y_proba, cv_scores):
    """Calculate and print all metrics for a model."""
    acc  = accuracy_score(y_true, y_pred)
    prec = precision_score(y_true, y_pred, zero_division=0)
    rec  = recall_score(y_true, y_pred, zero_division=0)
    f1   = f1_score(y_true, y_pred, zero_division=0)
    auc  = roc_auc_score(y_true, y_proba)
    cm   = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()

    print(f"  {'─'*55}")
    print(f"  MODEL: {name}")
    print(f"  {'─'*55}")
    print(f"  Accuracy  : {acc*100:.2f}%    (correct predictions / total)")
    print(f"  Precision : {prec*100:.2f}%    (of predicted defaults, how many are real)")
    print(f"  Recall    : {rec*100:.2f}%    (of real defaults, how many caught)")
    print(f"  F1-Score  : {f1*100:.2f}%    (harmonic mean of precision & recall)")
    print(f"  ROC-AUC   : {auc:.4f}     (rank discrimination ability)")
    print(f"  CV AUC    : {cv_scores.mean():.4f} ± {cv_scores.std():.4f}  (5-fold cross-validated)")
    print(f"\n  Confusion Matrix:")
    print(f"                   Predicted")
    print(f"                No Def  Default")
    print(f"  Actual No Def  [{tn:3d}]   [{fp:3d}]   ← False Positives (good rejected)")
    print(f"  Actual Default [{fn:3d}]   [{tp:3d}]   ← False Negatives (bad approved)")
    print(f"\n  Detailed Report:\n{classification_report(y_true, y_pred, target_names=['No Default','Default'])}")

    return {"name": name, "accuracy": acc, "precision": prec,
            "recall": rec, "f1": f1, "auc": auc, "cv_auc": cv_scores.mean(),
            "cv_std": cv_scores.std(), "cm": cm,
            "tn": tn, "fp": fp, "fn": fn, "tp": tp,
            "pred": y_pred, "proba": y_proba}
```

```python
# Evaluate all models
models_results = []
models_results.append(evaluate_model("Logistic Regression", y_test, lr_pred, lr_proba, lr_cv))
models_results.append(evaluate_model("Decision Tree",       y_test, dt_pred, dt_proba, dt_cv))
models_results.append(evaluate_model("Random Forest",       y_test, rf_pred, rf_proba, rf_cv))
models_results.append(evaluate_model("SVM (RBF Kernel)",    y_test, svm_pred, svm_proba, svm_cv))
```

```python
# Summary comparison table
print("\n" + "=" * 75)
print("  FINAL COMPARISON TABLE")
print("=" * 75)
print(f"  {'Model':<22} {'Accuracy':>9} {'Precision':>10} {'Recall':>8} {'F1':>7} {'AUC':>8} {'CV AUC':>10}")
print(f"  {'-'*70}")
for r in models_results:
    print(f"  {r['name']:<22} {r['accuracy']*100:>8.2f}% {r['precision']*100:>9.2f}% "
          f"{r['recall']*100:>7.2f}% {r['f1']*100:>6.2f}% {r['auc']:>8.4f} {r['cv_auc']:>8.4f}±{r['cv_std']:.3f}")
print("=" * 75)
```

```python
# Best model determination
best = max(models_results, key=lambda x: x['auc'])
print(f"\n  🏆 BEST MODEL (by ROC-AUC): {best['name']}  →  AUC = {best['auc']:.4f}")

```

```python
# ─────────────────────────────────────────────────────────
# STEP 6: FEATURE IMPORTANCE (Random Forest)
# ─────────────────────────────────────────────────────────
print("\n[STEP 6] Feature Importance Analysis (Random Forest)...")

feat_imp = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)
print(f"\n  {'Rank':<5} {'Feature':<12} {'Importance':>11} {'Contribution':>14}")
print(f"  {'-'*45}")
for rank, (feat, imp) in enumerate(feat_imp.items(), 1):
    bar = "█" * int(imp * 100)
    print(f"  {rank:<5} {feat:<12} {imp:>11.4f}  {bar}")

print(f"\n  Top 3 features account for "
      f"{feat_imp.head(3).sum()*100:.1f}% of predictive power")
```

```python
# ─────────────────────────────────────────────────────────
# STEP 7: CREDIT SCORE GENERATION (300–850 Scale)
# ─────────────────────────────────────────────────────────
print("\n[STEP 7] Generating Credit Scores...")

def prob_to_credit_score(prob_default, min_score=300, max_score=850):
    """
    Convert default probability to credit score.
    Formula: Score = 850 - (550 × P_default)
    Higher score = lower risk = better creditworthiness
    """
    return int(max_score - (max_score - min_score) * prob_default)

def credit_band(score):
    """Assign risk band based on credit score."""
    if score >= 750: return "Excellent  ✅"
    elif score >= 700: return "Good       ✅"
    elif score >= 650: return "Fair       ⚠️"
    elif score >= 600: return "Poor       🔶"
    else:              return "Very Poor  ❌"

# Build credit scores table (using best model = Logistic Regression for highest AUC)
score_table = X_test.copy().reset_index(drop=True)
score_table["Actual_Default"]    = y_test.values
score_table["LR_Prob"]           = lr_proba
score_table["RF_Prob"]           = rf_proba
score_table["Credit_Score_LR"]   = [prob_to_credit_score(p) for p in lr_proba]
score_table["Credit_Score_RF"]   = [prob_to_credit_score(p) for p in rf_proba]
score_table["Risk_Band"]         = [credit_band(s) for s in score_table["Credit_Score_LR"]]

print(f"\n  Credit Score Formula:")
print(f"  Score = 850 - (550 × P_default)  [Range: 300–850]")
print(f"\n  Risk Band Thresholds:")
print(f"  750–850 → Excellent  | 700–749 → Good | 650–699 → Fair")
print(f"  600–649 → Poor       | 300–599 → Very Poor")

print(f"\n  Sample Credit Scores (first 15 test records):")
display_cols = ["age", "income", "debtinc", "Actual_Default",
                "Credit_Score_LR", "Credit_Score_RF", "Risk_Band"]
print(score_table[display_cols].head(15).to_string(index=False))

band_dist = score_table["Risk_Band"].value_counts()
print(f"\n  Risk Band Distribution:")
print(band_dist.to_string())

# Save credit scores to CSV
score_table.to_csv("credit_scores_all_models.csv", index=False)
print(f"\n  ✓ Credit scores saved to: credit_scores_all_models.csv")

```
